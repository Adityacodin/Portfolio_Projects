```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('drive/My Drive/Projects/Central_Park_Squirrel_Project/2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20250326.csv')
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.loc[df["Specific Location"].isna(),"Specific Location"] = df.loc[df["Specific Location"].isna(),"Specific Location"].fillna("Unknown")
```

```python
df.loc[(df["Location"] == "Above Ground") & (df["Above Ground Sighter Measurement"].isna()),"Above Ground Sighter Measurement"] = df.loc[df["Above Ground Sighter Measurement"].isna(),"Above Ground Sighter Measurement"].fillna("Unknown")
```

```python
df.loc[(df["Above Ground Sighter Measurement"].isna()),["Location","Above Ground Sighter Measurement"]]
```

```python
df["Above Ground Sighter Measurement"].value_counts()
```

```python
df.loc[df["Above Ground Sighter Measurement"].isna(),"Above Ground Sighter Measurement"] = df.loc[df["Above Ground Sighter Measurement"].isna(),"Above Ground Sighter Measurement"].fillna("FALSE")
```

```python
df.loc[df["Location"]=="Above Ground","Above Ground Sighter Measurement"].value_counts()
```

```python
df.loc[df["Color notes"].isna(),"Color notes"] = df.loc[df["Color notes"].isna(),"Color notes"].fillna("No note")
```

```python
df.loc[df["Other Activities"].isna(),"Other Activities"] = df.loc[df["Other Activities"].isna(),"Other Activities"].fillna("No Other Activities")
df.loc[df["Other Interactions"].isna(),"Other Interactions"] = df.loc[df["Other Interactions"].isna(),"Other Interactions"].fillna("No Other Interactions")
```

```python
df.groupby("Age")["Primary Fur Color"].value_counts()
```

```python
df.loc[(df["Age"] == "Adult") & (df["Primary Fur Color"].isna()),"Age"].value_counts()
```

```python
df.loc[(df["Age"] == "Juvenile") & (df["Primary Fur Color"].isna()),"Age"].value_counts()
```

```python
df.loc[df["Primary Fur Color"].isna(),"Primary Fur Color"] = df.loc[df["Primary Fur Color"].isna(),"Primary Fur Color"].fillna(df["Primary Fur Color"].mode()[0])
```

```python
df.loc[df["Location"].isna(),"Location"] = df.loc[df["Location"].isna(),"Location"].fillna(df["Location"].mode()[0])
```

```python
df.loc[df["Age"].isna(),"Age"] = df.loc[df["Age"].isna(),"Age"].fillna(df["Age"].mode()[0])
```

```python
df.loc[df["Age"]=='?',"Age"] = df.loc[df["Age"]=='?','Age'].replace('?',df["Age"].mode()[0])
```

```python
df.loc[df["Primary Fur Color"].isna(),"Primary Fur Color"] = df.loc[df["Primary Fur Color"].isna(),"Primary Fur Color"].fillna(df.loc["Primary Fur Color"].mode()[0])
```

```python
df.groupby(["Hectare","Primary Fur Color"])["Highlight Fur Color"].value_counts()
```

```python
for index, row in df.iterrows():
  if pd.isna(row["Highlight Fur Color"]):
    try:
        mode_value = df.loc[(df["Hectare"] == row["Hectare"]) & (df["Primary Fur Color"] == row["Primary Fur Color"]),"Highlight Fur Color"].mode()
        if not mode_value.empty:
            mode_value = mode_value[0]
            df.loc[index, "Highlight Fur Color"] = mode_value
        else:
            df.loc[index, "Highlight Fur Color"] = "Unknown"
    except IndexError:
        pass
```

```python
df["Highlight Fur Color"].value_counts()
```

```python
filt_df = df.loc[df["Primary Fur Color"] == "Black"]
filt_df["Highlight Fur Color"].value_counts()
```

```python
df.loc[df["Highlight Fur Color"] == "Unknown","Highlight Fur Color"] = df.loc[df["Highlight Fur Color"] == "Unknown","Highlight Fur Color"].replace("Unknown","Cinnamon")
```

```python
df["Combination of Primary and Highlight Color"] = df["Primary Fur Color"] + "+" + df["Highlight Fur Color"]
```

```python
df["Combination of Primary and Highlight Color"].value_counts()
```
